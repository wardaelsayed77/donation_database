```python

query = """
SELECT d.Donation_Id, d.Don_Type_Id, dt.TypeName
FROM Donations d
JOIN DonationTypes dt ON d.Don_Type_Id = dt.Don_Type_Id
"""

donations = pd.read_sql(query, conn)


plt.figure(figsize=(7, 4))
sns.barplot(x=donations['TypeName'].value_counts().index, 
            y=donations['TypeName'].value_counts().values,color="lightgreen")

plt.xlabel('Type of donations')
plt.ylabel(' Number of donations')
plt.title('Distribution of donations by type')
plt.xticks(rotation=45) 
plt.show()
#conn.close()
```

```python
users = pd.read_sql("SELECT * FROM Users", conn)
donations = pd.read_sql("SELECT * FROM Donations", conn)
campaigns = pd.read_sql("SELECT * FROM Campaigns", conn)
requests = pd.read_sql("SELECT * FROM Requests", conn)
donation_types = pd.read_sql("SELECT * FROM DonationTypes", conn)
```

```python


campaign_donations = donations[donations['Camp_Id'].notnull()]
non_campaign_donations = donations[donations['Camp_Id'].isnull()]

plt.pie(
    [len(campaign_donations), len(non_campaign_donations)],
    labels=['Through campaigns', 'Individual donations'],
    autopct='%1.1f%%',colors=['lightgreen','darkseagreen']
)
plt.title('Campaign Donation Rate')
plt.show()
```

```python
donations['Donation_Date'] = pd.to_datetime(donations['Donation_Date'])
donations_by_month = donations.groupby(
    donations['Donation_Date'].dt.to_period('M')
)['Amount'].sum()

plt.figure(figsize=(4,4))
donations_by_month.plot(color="lightgreen")
plt.title('Monthly Donations ')
plt.xlabel('Mounth')
plt.ylabel('Amount')
plt.show()
```

```python

top_donors = donations.groupby('Donor_Id')['Amount'].sum().nlargest(6)
top_donors = top_donors.reset_index().merge(
    users, 
    left_on='Donor_Id', 
    right_on='UserId'
)

plt.figure(figsize=(10,5))
sns.barplot(data=top_donors, x='Name', y='Amount',color="lightgreen")
plt.title(' Top 6 Donors ')
plt.show()


top_acceptors = donations.groupby('Acceptor_Id')['Amount'].sum().nlargest(3)
top_acceptors = top_acceptors.reset_index().merge(
    users, 
    left_on='Acceptor_Id', 
    right_on='UserId'
)

plt.figure(figsize=(5,5))
sns.barplot(data=top_acceptors, x='Name', y='Amount',color="lightgreen")
plt.title('Top 3 Aceeptor  ')

plt.show()
```
